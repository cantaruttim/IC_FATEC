```python
from Bio.PDB import *
```

```python
pdbl = PDBList()
pdbl.retrieve_pdb_file('4XP1')
```

```python
parser = MMCIFParser()
```

```python
structure = parser.get_structure('4XP1', 'xp/4xp1.cif')
```

```python
def cleandir(obj):
    print(", ".join([a for a in dir(obj) if not a.startswith("_")]))
cleandir(structure)
```

```python
for model in structure:
    print(f"model {model}")
```

```python
model = structure[0] #since we only have one model
for chain in model:
    print(f"chain {chain}, Chain ID: {chain.id}")
```

```python
chain_A = model['A']
```

```python
for res in chain_A:
    print(f"Residue name: {res.resname}, number: {res.id[1]}")
```

```python
res = chain_A[56]
```

```python
print(res)
```

```python
for atom in res:
    print(f"{atom.name}")
```

```python
for model in structure:
    for chain in model:
        for residue in chain:
            for atom in residue:
                print(atom)
```
